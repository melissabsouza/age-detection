```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from roboflow import Roboflow
```

```python
rf = Roboflow(api_key="")
project = rf.workspace().project("age-detection-4ttq9")
model = project.version("2").model
```

```python
# infer on a local image
print(model.predict("/content/Captura de tela 2024-09-23 144953.png", confidence=40, overlap=30).json())
# visualize your prediction
model.predict("/content/Captura de tela 2024-09-23 144953.png", confidence=20, overlap=30).save("prediction2.jpg")
```
